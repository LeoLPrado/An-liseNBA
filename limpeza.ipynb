```python
import pandas as pd
```

```python
df = pd.read_csv('nba2k-full.csv')
display(df)
```

```python
# Tirando o '#' do numero das camisas

df['jersey'] = df['jersey'].str.replace('#', '')
display(df)
```

```python
# Pegando a altura em metros

df['height'] = df['height'].str[6:]
display(df)
```

```python
# Passando altura para float

df['height'] = df['height'].astype(float)
display(df.dtypes)
```

```python
# Removendo o '$' do salario

df['salary'] = df['salary'].str.replace('$', '')
display(df)
```

```python
# Passando salario para float

df['salary'] = df['salary'].astype(float)
```

```python
# Dropando algumas colunas

df.drop(columns = ['b_day', 'weight', 'college', 'version'], inplace = True)
```

```python
# Pegando o indice de linhas cujo o campo 'draft_round' seja 'Undrafted'

indexes = df.loc[df['draft_round'] == 'Undrafted'].index
df.drop(indexes , inplace=True)
display(df.shape)
```

```python
# Passando o rating para inteiro

df['rating'] = df['rating'].astype(int)
```

```python
display(df)
```

```python
# Pegando os salarios (min, max e a media) por posicao

df.groupby(['position']).agg({'salary': ['min', 'max', 'mean']})
```

```python
# Salario por regiao

df.groupby(['country']).agg({'salary': ['min', 'max']})
```

```python
# Passando os campos 'draft_round', 'draft_peak' e 'jersey' para inteiro para realizar analise visual

df['draft_round'] = df['draft_round'].astype(int)
df['draft_peak'] = df['draft_peak'].astype(int)
df['jersey'] = df['jersey'].astype(int)
display(df.dtypes)
```

```python
# Analises

import matplotlib.pyplot as plt
```

```python
# Gerando graficos dos campos abaixo

dados_numericos = df[['rating', 'jersey', 'height', 'salary', 'draft_year']]

for i in dados_numericos.columns:
    plt.hist(x = dados_numericos[i], color = 'blue')
    plt.title(i + ' Distribution', weight = 'bold')
    plt.show()
    print()
```

```python
import seaborn as sns
```

```python
# Grafico para analisar salario em funcao do rating

sns.scatterplot(x = df['rating'], y = df['salary'], color = 'red')
plt.show()
```

```python
# Graficos dos campos abaixo

dados_string = df[['team', 'country', 'position']]

for i in dados_string.columns:
    plt.figure(figsize = (12,6))                            # Definindo tamanho
    plt.title(i + ' Distribution', weight = 'bold')         # Definindo formato de barras e nome do grafico
    sns.countplot(x = dados_string[i], palette = 'Set1')    # Definindo o eixo x
    plt.xticks(rotation = 90)                               # rotacao da legenda no eixo x
    plt.show()
    print()
```
